```python
import requests
from bs4 import BeautifulSoup
from tkinter import *
from tkinter.messagebox import *


def main():
    url = 'http://www.cdc.go.kr/linkCheck.es?mid=a21111050500'
    response = requests.get(url)
    dom_a = BeautifulSoup(response.content, 'html.parser')

    update_time = dom_a.select_one(".content_body > p").text.strip()
    patient_total = dom_a.select_one(".content_body > ul > li").text.strip()
    
    root = Tk()
    root.title("Corona system")
    root.geometry("500x400+100+100")
    root.resizable(False, False)
    frame= Frame(root)
    label1 = Label(frame, text= update_time[21:])
    label2 = Label(frame, text= "감염 현황 : " + patient_total)
    button1 = Button(frame, text="update")

    label1.grid(row=0,column=0, sticky='ew')
    label2.grid(row=1,column=0, sticky='ew')
    button1.grid(row=2,column=0)
    frame.pack(fill="both")

    def btn_click(event):
        url = 'http://www.cdc.go.kr/linkCheck.es?mid=a21111050500'
        response = requests.get(url)
        dom_a = BeautifulSoup(response.content, 'html.parser')

        update_time = dom_a.select_one(".content_body > p").text.strip()
        patient_total = dom_a.select_one(".content_body > ul > li").text.strip()
        label1.config(text = update_time[21:])
        label2.config(text="감염 현황 : "+ patient_total)

    button1.bind('<Button-1>', btn_click)
    
    root.mainloop()
main()
```

```python
url2 = 'http://www.cdc.go.kr/linkCheck.es?mid=a21111050200'

response2 = requests.get(url2)
dom_a2 = BeautifulSoup(response2.content, 'html.parser')
link_raw = dom_a2.select(".mt30 > ul > li")
links = [link.find('a')['href'] for link in link_raw]

doms = []
for link in links:
    response3 = requests.get(link)
    dom_a3 = BeautifulSoup(response3.content,'html.parser')
    doms.append(dom_a3)
print(doms[0])
```

```python
import pandas as pd
from datetime import datetime

pub_dates = [datetime.strptime(link.text.split('[')[1].split(']')[0],'%Y.%m.%d') for link in link_raw]
print(pub_dates[0]," type : ",type(pub_dates[0]))

df_bodo = pd.DataFrame({'pub_date':pub_dates, 'dom':doms,'link':links})
df_bodo.to_csv('corona_bodo_20200203.csv',index=False)
```
